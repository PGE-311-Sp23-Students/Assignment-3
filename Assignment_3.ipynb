In Problem 2 you will practice using a fun linear algebra application, the rotation matrix.
The process of rotating an object with respect to an angle in a two-dimensional plane is 2D rotation. We accomplish this rotation with the help of a 2 x 2 rotation matrix that has the standard form as given below:
$$
M(\theta)=\left[\begin{matrix}
cos(\theta) & -sin(\theta)\\
sin(\theta) & cos(\theta)
\end{matrix}\right]
$$
If we want to rotate a vector with the coordinates (x, y) then we use matrix multiplication to perform the rotation as follows: 
$$
\left[\begin{matrix}
x_1^{new}   & y_1^{new}\\
\cdots & \cdots\\
x_n^{new}   & y_n^{new}
\end{matrix}\right]=\left[\begin{matrix}
x_1   & y_1\\
\cdots & \cdots\\
x_n   & y_n
\end{matrix}\right]\left[\begin{matrix}
cos(\theta) & -sin(\theta)\\
sin(\theta) & cos(\theta)
\end{matrix}\right]
$$

#### e) Plot isosurface contour at the level=0.1 using pyvista. 
Install packages in your prompt shell or terminal to enable interactive plotting in Jupyter notebooks.  >>pip install pythreejs

```python
import pyvista as pv

pv.set_jupyter_backend('pythreejs')

def initialize_plotter():
      
    plotter_obj = pv.Plotter()
      
    plotter_obj.set_background(color='w')
    pv.global_theme.font.color = 'black'
    return plotter_obj

pv_img = pv.wrap(data)
```

```python

p = initialize_plotter()

# Generate a contoured surface at the explicitly specified isosurface value. 
contour = pv_img.contour(isosurfaces=[])
 
p.add_mesh(contour, color=[0, 0.8, 0.27], opacity=1.)

p.show()
```
